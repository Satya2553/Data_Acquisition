Writing a Function to calculate mode

```python
def mode(c):
  f={}
  for n in c:
    f.setdefault(n,0)
    f[n]+=1
    h=max(f.values())
    mode=[]
    for m,x in f.items():
        if x==h:
          mode.append(m)
  return mode
```

```python
print(mode(c))
```
